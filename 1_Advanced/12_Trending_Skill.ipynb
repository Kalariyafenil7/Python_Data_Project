```python
import pandas as pd
import ast
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply({lambda x: ast.literal_eval(x) if pd.notna(x) else(x)})
```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst'].copy()

df_DA
```

```python
df_DA['job_posted_month_No.'] = df_DA['job_posted_date'].dt.month
```

```python
df_DA
```

```python
df_DA_explode = df_DA.explode('job_skills')
```

```python
df_DA_explode
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month_No.', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
df_DA_pivot
```

```python
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]

df_DA_pivot.drop('Total', inplace=True)

df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(kind='line')
plt.title('Top 5 Trending Skills for Data Analysts Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.legend(title='Skills')
plt.show()
```

```python
df_DA_pivot1 = df_DA_pivot.reset_index()
df_DA_pivot1['job_posted_month'] = df_DA_pivot1['job_posted_month_No.'].apply(lambda x: pd.to_datetime(x, format="%m").strftime('%b'))
df_DA_pivot1 = df_DA_pivot1.set_index('job_posted_month')
df_DA_pivot1 = df_DA_pivot1.drop(columns=['job_posted_month_No.'])

df_DA_pivot1.iloc[:, :5].plot(kind='line')
plt.title('Top 5 Trending Skills for Data Analysts Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.legend(title='Skills')
plt.ylim(0, 12500)
plt.show()
```
