```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
df
```

```python
months = df['job_posted_month'].unique()

months_dics = {month: df[df['job_posted_month'] == month] for month in months}

months_dics['Jan']

```

```python
df_Q1 = pd.concat([months_dics['Jan'],months_dics['Feb'],months_dics['Mar']],ignore_index=True)
```

```python
df_Q1['job_posted_month'].value_counts().plot(kind='bar')
plt.title('Job Postings in Q1')
plt.xlabel('Month')
plt.ylabel('Number of Job Postings')
plt.xticks(rotation=0)
```
