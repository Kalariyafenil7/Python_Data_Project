```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.pivot_table(index='job_title_short',aggfunc='size')
```

```python
df.groupby('job_title_short')['job_title_short'].count()
```

```python
df.pivot_table(index='job_title_short', values='salary_year_avg',columns='job_country', aggfunc='median')
```

```python
top_countries = df['job_country'].value_counts().head(6).index

top_countries 
```

```python
df_jobs_country_salary = df.pivot_table(index='job_country', values='salary_year_avg',columns='job_title_short', aggfunc='median')
df_jobs_country_salary = df_jobs_country_salary.loc[top_countries]

job_titles = ['Data Analyst','Data Scientist','Data Engineer']
df_jobs_country_salary = df_jobs_country_salary[job_titles]

df_jobs_country_salary.plot(kind='bar')
plt.title('Job Salary by Job Title and Country')
plt.ylabel('Salary (USD)')
plt.xlabel('Country')
plt.xticks(rotation=45, ha = 'right')
```
