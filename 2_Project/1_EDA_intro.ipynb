```python
import pandas as pd
import ast
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply({lambda x: ast.literal_eval(x) if pd.notna(x) else(x)})
```

Filter for India Data Analyst Job Role

```python


df_DA_india = df[(df['job_title_short'] == 'Data Analyst') & (df['job_location']=='India')].copy()

```

```python

df_DA_US = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')]
df_DA_india = df[(df['job_country'] == 'India') & (df['job_title_short'] == 'Data Analyst')]
```

```python
df_plot =df_DA_india['job_location'].value_counts().head(10).to_frame()

sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:blue_r')
sns.set_style('ticks')
sns.despine()
```

```python
dict_columns = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'No Degree Mention',
    'job_health_insurance': 'Health Insurance'
}
fig, ax = plt.subplots(1, 3)
fig.set_size_inches(15, 5)
for i, (column,title) in enumerate(dict_columns.items()):
    ax[i].pie(df_DA_US[column].value_counts(), startangle=90, autopct='%1.1f%%', labels=['False', 'True'])
    ax[i].set_title(title)
    ax[i].set_ylabel('')
    ax[i].set_xlabel('')
    ax[i].set_aspect('equal')
    ax[i].legend(loc='upper right', labels=['False', 'True'])
```

```python
df_plot =df_DA_india['company_name'].value_counts().head(10).to_frame()

sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:blue_r')
sns.set_style('ticks')
sns.despine()
plt.title('Top 10 Companies Hiring Data Analysts in India')
plt.xlabel('Number of Job Postings')
plt.ylabel('Company Name')
plt.show()
```
